```python
from docent import Docent
import pandas as pd

API_KEY = "dk_hfpusaj97r9Ix9yP_CdhCmvLQXpmLbi4lipuwgq6XnhTMcHzyBLYncESe8yrl0B"
SERVER_URL = "http://localhost:8890"
COLLECTION_ID = "7d99d911-8bfd-44f1-bb06-ca4ad59c5c3e"

DQL_QUERY = """
SELECT
  ar.id AS agent_run_id,
  ar.metadata_json->>'agent' AS metadata_agent,
  ar.metadata_json->'agent'->'extra'->>'parser' AS metadata_agent_extra_parser,
  ar.metadata_json->'agent'->'extra'->>'temperature' AS metadata_agent_extra_temperature,
  ar.metadata_json->'agent'->>'model_name' AS metadata_agent_model_name,
  ar.metadata_json->'agent'->>'name' AS metadata_agent_name,
  ar.metadata_json->'agent'->>'version' AS metadata_agent_version,
  ar.metadata_json->>'atif_session_id' AS metadata_atif_session_id,
  ar.metadata_json->>'eval_name' AS metadata_eval_name,
  ar.metadata_json->>'exception' AS metadata_exception,
  ar.metadata_json->>'final_metrics' AS metadata_final_metrics,
  ar.metadata_json->'final_metrics'->>'total_cached_tokens' AS metadata_final_metrics_total_cached_tokens,
  ar.metadata_json->'final_metrics'->>'total_completion_tokens' AS metadata_final_metrics_total_completion_tokens,
  ar.metadata_json->'final_metrics'->>'total_cost_usd' AS metadata_final_metrics_total_cost_usd,
  ar.metadata_json->'final_metrics'->>'total_prompt_tokens' AS metadata_final_metrics_total_prompt_tokens,
  ar.metadata_json->>'job_id' AS metadata_job_id,
  ar.metadata_json->>'scores' AS metadata_scores,
  ar.metadata_json->'scores'->>'reward' AS metadata_scores_reward,
  ar.metadata_json->>'task' AS metadata_task,
  ar.created_at AS created_at
FROM agent_runs ar
ORDER BY ar.created_at ASC
"""

client = Docent(api_key=API_KEY, server_url=SERVER_URL)
result = client.execute_dql(COLLECTION_ID, DQL_QUERY)
df = client.dql_result_to_df_experimental(result)

df.head()
```
