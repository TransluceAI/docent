```python
from docent import DocentClient
client = DocentClient(server_url="http://localhost:8888", web_url="http://localhost:3002")
```

```python
from docent._loader.load import load_picoctf_4o, load_picoctf_36, load_agentharm, load_cybench, load_tau_bench, load_oh_swe_bench
from docent._frames.types import Datapoint

datasets = [
    # ("picoCTF GPT-4o", load_picoctf_4o),
    # ("picoCTF sonnet 3.5", load_picoctf_36),
    # ("agentHarm", load_agentharm),
    # ("tau-bench", load_tau_bench),
    # ("oh-swe-bench", load_oh_swe_bench),
    ("cybench", load_cybench),
]

for name, load_fn in datasets:
    fg_id = client.create_framegrid(name=name)
    transcripts = load_fn()
    client.add_datapoints(fg_id, [Datapoint.from_transcript(t) for t in transcripts])
```
